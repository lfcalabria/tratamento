```python
dias = [i for i in range(1, 32)]
apac = apac.melt(
    id_vars=['Código', 'Posto', 'Mês/Ano'],
    value_vars=dias,
    var_name='Dia',
    value_name='Chuva_mm_str'
)
apac
```

```python
apac['Data'] = pd.to_datetime(
    apac['Mês/Ano'].dt.strftime('%Y-%m-') + apac['Dia'].astype(str),
    errors='coerce'
)
apac
```

```python
apac.dropna(subset=['Data'], inplace=True)
apac
```

```python
apac['Chuva_mm'] = pd.to_numeric(apac['Chuva_mm_str'], errors='coerce')
apac
```

```python
apac = apac.groupby(['Data', 'Código'])['Chuva_mm'].sum().reset_index()
apac
```

```python
apac = apac.pivot_table(
    index='Data',
    columns='Código',
    values='Chuva_mm'
)
apac
```

```python
apac.info()
```

```python
min_date = apac.index.min()
max_date = apac.index.max()
idx = pd.date_range(start=min_date, end=max_date, freq='D')
apac = apac.reindex(idx, fill_value=0)
apac.info()
```
